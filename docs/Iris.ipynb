# Structure Figure on Iris

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import datasets
from numpy.random import permutation
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target

# Randomize samples
p = permutation(iris_X.shape[0])
iris_X = iris_X[p]
iris_y = iris_y[p]

fig = plt.figure(figsize=(30, 15))

# Update the tree and plot tree structures
sdt = DecisionTreeClassifier()
sdt.partial_fit(iris_X[:25], iris_y[:25], classes=[0, 1, 2])
sdt_plot = plot_tree(
    sdt,
    feature_names=iris.feature_names,
    class_names=iris.target_names.tolist(),
    rounded=True,
    filled=True,
    ax=fig.add_axes([0, 0.3, 0.3, 0.7]),
)
sdt.partial_fit(iris_X[25:50], iris_y[25:50])
sdt_plot = plot_tree(
    sdt,
    feature_names=iris.feature_names,
    class_names=iris.target_names.tolist(),
    rounded=True,
    filled=True,
    ax=fig.add_axes([0.3, 0, 0.7, 1]),
)
plt.savefig("../paper/iris.pdf", transparent=True, bbox_inches="tight")
```

```python
sdt.predict(iris_X) == iris_y
```
